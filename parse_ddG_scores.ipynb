```python
import json
```

```python
data_file = open('total_OMPF_DDGun.out')

data_dict = {}
wt_sequence = {}

for line in data_file:
    if line.startswith('#'):
        header_items = line[1:].strip().split()
    else:
        items = line.strip().split()
        chain = items[1]
        mutation = items[2]
        wtaa = mutation[0]
        mutaa = mutation[-1]
        resnum = int(mutation[1:-1])
        score = float(items[3])
        wt_sequence[resnum] = wtaa
        if resnum in data_dict:
            data_dict[resnum][mutaa] = score
        else:
            data_dict[resnum] = {mutaa:score,wtaa:0.0}
```

```python
wt_sequence
```

```python
f = open('2omf_wt_sequence_dict.json','w')
f.write(json.dumps(wt_sequence,indent=True))
f.close()
```

```python
f = open('2omf_ddgun.json','w')
f.write(json.dumps(data_dict,indent=True))
f.close()
```
