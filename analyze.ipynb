```python
import os
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
INPUT_DIR = './inputs/3_conformers/best-scores'
input_files = [INPUT_DIR+'/'+f for f in os.listdir(INPUT_DIR)]
```

```python
def _get_second(a): return a[1]
```

```python
SIMULATION_DIR = './simulations/trial6'
MAINDIR = os.getcwd()

scores_models_w_hbonds = []
scores_models = []

for input_file in input_files:
    label = input_file.split('/')[-1][:-4]
    os.chdir('{0}/{1}'.format(SIMULATION_DIR,label))
    scorefiles = [each for each in os.listdir('.') if each.endswith('.sc')]
    for scorefile in scorefiles:
        df = pd.read_fwf(scorefile,header=1,index_col=-1)
        sc = df['if_X_hbond_sc']
        lrbb = df['if_X_hbond_lr_bb']
        bbsc = df['if_X_hbond_bb_sc']
        srbb = df['if_X_hbond_sr_bb']
        for s,l,b,sr,i,index in zip(sc,lrbb,bbsc,srbb,df['interface_delta_X'],df.index):
            filepath = '{0}/{1}/{2}.pdb'.format(SIMULATION_DIR,label,index)
            scores_models.append([filepath,i])
            if s<0 or l<0 or b<0 or sr<0:
                scores_models_w_hbonds.append([filepath,i])
    os.chdir(MAINDIR)
```

```python
scores_models_w_hbonds_sorted = sorted(scores_models_w_hbonds,key=_get_second)
scores_models_sorted = sorted(scores_models,key=_get_second)
```

```python
def compute_ddGun_score(sequence_dict,ddg_score_dict='../../../provean_DDGun_results/2omf_ddgun.json'):
    ddg_data = json.loads(open(ddg_score_dict).read())
    score_dict = {}
    #print(ddg_data.keys())
    for resi, aa in sequence_dict.items():
        resi = str(resi)
        score_dict[resi] = ddg_data[resi][aa]
#     print(ddg_data[resi][aa])
    scores = list(score_dict.values())
    for key in score_dict:
        if score_dict[key]>0:
            print(key,score_dict[key])
    max_score = max(scores)
    min_score = min(scores)
    avg_score = sum(scores)/len(scores)
#     print(max_score,min_score)
    return (max_score+min_score-avg_score)
```
